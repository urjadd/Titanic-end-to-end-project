### Tickets

```python
TICKETS_SPLITTED_TRAIN=train1['Ticket'].str.replace('.','').str.replace('/','').str.split().to_numpy()
c_dict_train = {
    'OTHER' : 0
}

ticket_identifier_column = []
for e in TICKETS_SPLITTED_TRAIN:
        if len(e) == 1:
            c_dict_train['OTHER'] += 1
            ticket_identifier_column.append('OTHER')
        elif e[0] not in c_dict_train: 
            c_dict_train[e[0]] = 1
            ticket_identifier_column.append(e[0])
        else: 
            ticket_identifier_column.append(e[0])
            c_dict_train[e[0]] += 1

train1['IdentifierTicket'] = ticket_identifier_column
```
